```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms
```

Tuned police result

```python
tuned_police_episode = 499
tuned_police_folder = 'C:\\Users\\Asus\\OneDrive - Chulalongkorn University\\Desktop\\Senior Project\\Experiments-with-DQN-Onnut-27042022\\Experiments-with-DQN-Onnut-21042022\\multi-agent-dqn-main\\tuned_police_{}episodes\\tuned_police'.format(tuned_police_episode)

reward_onnut = []
reward_virtual = []

travel_time_onnut = []
travel_time_virtual = []

tuned_police_index = []

for i in range(tuned_police_episode) :
    tuned_police_csv_path = os.path.join(tuned_police_folder,'tuned_police_output_run{}.csv'.format(i))
    tuned_police_df = pd.read_csv(tuned_police_csv_path)

    tuned_police_index.append(i)

    sum_reward_onnut = tuned_police_df['reward_onnut'].sum()
    mean_reward_onnut = tuned_police_df['reward_onnut'].mean()
    # reward_onnut.append((sum_tuned_police_reward_onnut,mean_tuned_police_reward_onnut))
    reward_onnut.append(sum_reward_onnut)

    sum_reward_virtual = tuned_police_df['reward_virtual'].sum()
    meane_reward_virtual = tuned_police_df['reward_virtual'].mean()
    # reward_virtual.append((sum_tuned_police_reward_virtual,mean_tuned_police_reward_virtual))
    reward_virtual.append(sum_reward_virtual)

    sum_travel_time_onnut = tuned_police_df['travel_time_onnut'].sum()
    mean_travel_time_onnut = tuned_police_df['travel_time_onnut'].mean()
    # travel_time_onnut.append((sum_tuned_police_travel_time_onnut,mean_tuned_police_reward_onnut))
    travel_time_onnut.append(mean_travel_time_onnut)

    sum_travel_time_virtual = tuned_police_df['travel_time_virtual'].sum()
    mean_travel_time_virtual = tuned_police_df['travel_time_virtual'].mean()
    travel_time_virtual.append(mean_travel_time_virtual)


sum_tuned_police_reward_onnut = np.mean(reward_onnut)
sum_tuned_police_reward_virtual = np.mean(reward_virtual)
mean_tuned_police_travel_time_onnut = np.mean(travel_time_onnut)
mean_tuned_police_travel_time_virtual = np.mean(travel_time_virtual)
```
